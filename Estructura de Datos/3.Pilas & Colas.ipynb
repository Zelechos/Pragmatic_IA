# Pilas & Colas

# Pilas

```python
class Pila():
    
    #Metodo Constructor 
    def __init__(self,Size):
        self.Lista = []
        self.Tope = 0
        self.Size = Size
    
    #Metodo que Verifica si la Pila esta vacia
    def empty(self):
        if self.Tope == 0:
            print('Pila vacia')
            return True
        else:
            print('Pila tiene elementos')
            return False
    
    #Metodo para agregar un dato a la Pila
    def push(self,dato): 
        if self.Tope < self.Size:
            self.Lista += [dato] 
            self.Tope += 1
            print('Se Agrego el Elemento ',dato,' a la Pila')
        else:
            self.Size += 5            
            self.Lista += [dato]
            self.Tope += 1
            print('La Pila esta llena.')
    
    #Metodo para eliminar un dato a la Pila
    def pop(self):
        if self.empty() == True:
            print('No hay Elementos en la Pila')
        else:
            self.Tope -= 1
            self.Lista = [self.Lista[x] for x in range(self.Tope)]
            print('Se Elimino un Elemento')
            
    #Metodo para mostrar los datos que estan en la Pila 
    def showMe(self):
        if self.Tope == 0:
            print('Pila Vacia')
        else:
            print('Pila Actual:')
        
        for index in range(self.Tope):
            print('[',index,'] => ',self.Lista[index])

    #Metodo para mostrar el tamaÃ±o de nuestra Pila
    def size(self):
        return print('La Cantidad de Elementos Actual de la Pila es: ',self.Tope)
    
    #Metodo Para mostrar el ultimo elemento de la Pila 
    def top(self):
        if self.Tope == 0:
            return print('La Pila Esta Vacia')
        return print('El Ultimo Elemento de la Pila es: ',self.Lista[-1])
    
```

```python
pila = Pila(1)

while True:   
    op = int(input("********MENU**********\n 0.Mostrar Pila\n 1.Agregar Dato\n 2.Borrar Dato\n 3.Vacio \n 4.Cantidad de Datos\n 5.Primer Dato \n 6.Salir \n Digite una Opcion del Menu :"))
    if op is 0:
        print('---------------------------------------------------------------------------------------------')
        pila.showMe()
        print('---------------------------------------------------------------------------------------------')
    elif op is 1:    
        print('---------------------------------------------------------------------------------------------')
        pila.push(int(input('Digite dato Que va Agregar:')))
        print('---------------------------------------------------------------------------------------------')
    elif op is 2:    
        print('---------------------------------------------------------------------------------------------')
        pila.pop()
        print('---------------------------------------------------------------------------------------------')
    elif op is 3:
        print('---------------------------------------------------------------------------------------------')
        pila.empty()
        print('---------------------------------------------------------------------------------------------')
    elif op is 4:
        print('---------------------------------------------------------------------------------------------')
        pila.size()
        print('---------------------------------------------------------------------------------------------')
    elif op is 5:
        print('---------------------------------------------------------------------------------------------')
        pila.top()
        print('---------------------------------------------------------------------------------------------')
    elif op is 6:
        print('---------------------------------------------------------------------------------------------')
        print('Fin del Programa. . .')
        print('---------------------------------------------------------------------------------------------')
        break
    else:
        print('La Opcion que selecciono no existe intente de nuevo')


```

# Colas

```python
class Cola():
    
    def __init__(self):
        self.cola = []
        self.size = 0
        
    def empty(self):
        return len(self.cola) ==  0 
    
    def push(self,Data):
        self.cola += [Data]
        self.size += 1
    
    def pop(self):
        if self.empty() == True:
            print('La Cola esta vacia.')
        else:
            self.cola = [self.cola[x]  for x in range(1,self.size)]
            self.size -= 1
            
    def show(self):
        n = self.size - 1
        while n > -1:
            print('[',n,'] => ',self.cola[n])
            n -= 1
    
    def front(self):
        if self.empty == True:
            print('La Cola esta vacia.')
        else:   
            print('El primer dato la Cola es : ',self.cola[0])
```

## Ejercicio 0 
con una Pila crear un estante de discos de peliculas que :
- Hacer un Menu con la siguientes funcionalidades
- Introduzca nuevos discos
- Elimine el ultimo disco introducido
- Que muestre todas las Peliculas
- Que muestre la Primer Pelicula Introducida
- Que muestre la Ultima Pelicula Introducida
- Que muestre la cantidad de Discos Actual

```python
class Discos():
    
    def __init__(self, Size):
        self.Estante = []
        self.Tope = 0
        self.Size = Size
    
    def Empty(self):
        if self.Tope == 0:
            print("El estante de Discos esta Vacio")
            return True
        else:
            print("El estante de Discos Tiene peliculas")
            return False
        
    def Push(self, Data):
        if self.Tope < self.Size:
            self.Estante += [Data]
            self.Tope += 1
            print("Se Agrego la Pelicula ",Data," al Estante de Discos...")
        else:
            self.Size += 5
            self.Estante += [Data]
            self.Tope += 1
            print("El estante esta lleno")
            
    def PoP(self):
        if self.Empty == True:
            print("El estante de Discos esta Vacio")
        else:
            self.Tope -= 1
            self.Estante = [self.Estante[x] for x in range(self.Tope)]
            print("Se elimino un Disco del Estante . . .")
            
    def ShowMe(self):
        if self.Tope == 0:
            print("El estante de Discos esta Vacio")
        else:
            print("Contenido del Estante: ")
            for Disco in range(len(self.Estante)):
                print("[Disco",Disco,"]====>  "+ self.Estante[Disco])
                
    def All(self):
        if self.Empty == True:
            print("La Cantidad de Discos Actual es : 0 ")
        else:
            print("La Cantidad de Discos Actual es : ", self.Tope)
            
    def LastMovie(self):
        if self.Tope == 0:
            print("El estante de Discos esta Vacio")
        else:
            print("La Ultima Pelicula introducida es : "+ self.Estante[-1])
        
    def FirstMovie(self):
        if self.Tope == 0:
            print("El estante de Discos esta Vacio")
        else:
            print("La Primera Pelicula introducida es : "+ self.Estante[0])
        
```

```python
Estante = Discos(5)

while True:
    Opcion = int(input("""
                          ***************************Menu***************************
                          |      {0}.- Insertar nuevo Disco.                       | 
                          |      {1}.- Elimine el Ultimo Disco insertado.          |
                          |      {2}.- Ver Todas la Peliculas.                     |
                          |      {3}.- Ver la Primera Pelicula insertada.          |
                          |      {4}.- Ver la Ultima Pelicula insertada.           |
                          |      {5}.- Ver la Cantidad de Peliculas Actual.        |
                          |      {6}.- Salir.                                      |
                          **********************************************************
         Elige un Opcion : ~"""))
    
    #Estructura switch
    if Opcion == 0 :
        Estante.Push(input("Introduzca el Nombre de la Pelicula : "))
    elif Opcion == 1:
        Estante.PoP()
    elif Opcion == 2:
        Estante.ShowMe()
    elif Opcion == 3:
        Estante.FirstMovie()
    elif Opcion == 4:
        Estante.LastMovie()
    elif Opcion == 5:
        Estante.All()
    elif Opcion == 6:
        print("Fin del Programa . . .")
        break
    
```
